##**Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca file CSV
data = pd.read_csv('Jumlah Penduduk Miskin, 2023.csv')

# Melihat sekilas data
print(data)
```

Data yang digunakan adalah jumlah penduduk miskin (ribu jiwa) per wilayah di Provinsi Lampung tahun 2023. Data tersebut memuat 15 wilayah yang berbeda dan jumlah penduduk miskin untuk setiap wilayah. Data tersebut didapat dari Badan Pusat Statistik (BPS) Lampung.

##**Visualisasi**

```python
# Membuat visualisasi barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Jumlah Penduduk Miskin (ribu jiwa)', y='Wilayah', data=data, palette='Blues_d')

# Menambahkan judul dan label
plt.title('Jumlah Penduduk Miskin di Berbagai Wilayah Lampung (ribu jiwa)', fontsize=16)
plt.xlabel('Jumlah Penduduk Miskin (ribu jiwa)', fontsize=12)
plt.ylabel('Wilayah', fontsize=12)

# Menampilkan visualisasi
plt.tight_layout()
plt.show()
```

##**Tujuan, User, Fungsi, dan Tone:**
- Tujuan: Menyajikan distribusi jumlah penduduk miskin di berbagai wilayah di Lampung tahun 2023 secara visual untuk memudahkan pembuat kebijakan atau masyarakat dalam memahami wilayah mana yang memiliki tingkat kemiskinan tertinggi.

- User: Pembuat kebijakan, peneliti sosial, pemerintah daerah, atau masyarakat umum yang tertarik dengan data kemiskinan di Provinsi Lampung.

- Fungsi: Visualisasi ini bertujuan untuk menunjukkan perbandingan jumlah penduduk miskin di berbagai wilayah di Lampung pada tahun 2023 agar pengguna dapat melihat wilayah dengan jumlah kemiskinan yang tinggi dan membuat keputusan yang tepat dalam alokasi bantuan atau program pengentasan kemiskinan.

- Tone: Informatif dan profesional. Visualisasi jelas, mudah dipahami, dan tidak terlalu mencolok, fokus pada penyajian data yang akurat dan mudah diinterpretasi.
